## Problem 1.2

### (a)

```python
data = pd.get_dummies(data, columns = ['device', 'light', 'motion'])
```

```python
data.head(10)
```

### (b)

```python
std_df = data
#Defining the variables to normalize
std_vars = ["co", "humidity", "lpg", "smoke", "temp"]
#Normalizing the data using (df-df.mean())/df.std()
#Indexing the data for each device
std_df[std_df['device_00:0f:00:70:91:0a'] == 1] = std_df[std_df['device_00:0f:00:70:91:0a'] == 1].apply(lambda x: (x - x.mean(skipna=True))/x.std(skipna=True) if x.name in std_vars else x)
std_df[std_df['device_1c:bf:ce:15:ec:4d'] == 1] = std_df[std_df['device_1c:bf:ce:15:ec:4d'] == 1].apply(lambda x: (x - x.mean(skipna=True))/x.std(skipna=True) if x.name in std_vars else x)
std_df[std_df['device_b8:27:eb:bf:9d:51'] == 1] = std_df[std_df['device_b8:27:eb:bf:9d:51'] == 1].apply(lambda x: (x - x.mean(skipna=True))/x.std(skipna=True) if x.name in std_vars else x)
```

### (e)

```python
copy = pd.read_csv('datasets/iot_telemetry_dataset.csv')
means_df = copy.groupby(['device']).mean().reset_index()
vars_df = copy.groupby(['device']).var().reset_index()

means_df = means_df[["co", "humidity", "lpg", "smoke", "temp", "device"]]
vars_df = vars_df[["co", "humidity", "lpg", "smoke", "temp", "device"]]

means_df = means_df.drop("device", 1)
means_df = means_df.transpose()
vars_df = vars_df.drop("device", 1)
vars_df = vars_df.transpose()

mean_plot = means_df.plot.bar(rot = 0)
vars_plot = vars_df.plot.bar(rot = 0)
```

Plotting all five variables in the same plot, causes a scale issue as some variables have high means and variances, so the means and variances of the other variables (co, lpg, and smoke) are undiscernable. Also notice that the variance for device 1 (1c:bf:ce:15:ec:4d) is a lot higher than the other devices. This is especially noticeable in the variance of humidity.

## Problem 1.3

```python
mapped_vars = ['co','humidity','lpg','smoke','temp']
std_df['mean'] = std_df[mapped_vars].mean(axis = 1)
std_df['variances'] = std_df[mapped_vars].var(1)
std_df['kurtosis'] = sps.kurtosis(std_df[mapped_vars], axis = 1)
std_df['skew'] = sps.skew(std_df[mapped_vars], axis = 1)
std_df = std_df.drop(labels=['co','humidity','lpg','smoke','temp'],axis=1)
std_df.head(40)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(121, projection = "3d")
ax1.scatter3D(matrix_X[:,0], matrix_X[:,1], matrix_X[:,2], color = "magenta")

ax2 = fig.add_subplot(122, projection = "3d")
ax2.scatter3D(matrix_X_hat2[:,0], matrix_X_hat2[:,1], matrix_X_hat2[:,2], color = "cyan")
```

The shape of the scatterplot for the reconstructed values is a line. This is due to using only one principal component to project x. This causes the projection to use k = 1, reducing the dimension to 1. When reconstructing x from the projection, there is no way to also reconstruct the reduced dimensions, leaving the reconstructed x to only be in one dimension, a line.
